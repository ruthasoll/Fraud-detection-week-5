```python
# Cell 1: Imports (same as above)
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Cell 2: Load data
df_credit = pd.read_csv('../data/raw/creditcard.csv')

# Cell 3: Overview
print("Shape:", df_credit.shape)
df_credit.head()

# Cell 4: Missing values & duplicates
print(df_credit.isnull().sum())
df_credit.drop_duplicates(inplace=True)
print("After removing duplicates - Shape:", df_credit.shape)

# Cell 5: Save cleaned
df_credit.to_csv('../data/processed/cleaned_creditcard.csv', index=False)

# Cell 6: Univariate - Amount and Time
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
sns.histplot(df_credit['Amount'], bins=50, kde=True, ax=axes[0])
axes[0].set_title('Transaction Amount Distribution')
sns.histplot(df_credit['Time'], kde=True, ax=axes[1])
axes[1].set_title('Time Distribution')
plt.tight_layout()
plt.show()

# Cell 7: Bivariate - Amount vs Class
plt.figure(figsize=(8,5))
sns.boxplot(x='Class', y='Amount', data=df_credit)
plt.title('Amount vs Fraud Class')
plt.show()

# Cell 8: Class imbalance
plt.figure(figsize=(6,4))
sns.countplot(x='Class', data=df_credit)
plt.title('Class Distribution - Credit Card')
plt.show()

fraud_rate = df_credit['Class'].mean() * 100
print(f"Fraud rate: {fraud_rate:.4f}% (extremely imbalanced)")
```
